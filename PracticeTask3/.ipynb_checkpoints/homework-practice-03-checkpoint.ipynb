```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.preprocessing import StandardScaler, MinMaxScaler
def pairplot(df, target, ysign='target', scale=4):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * scale, nrow * scale))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature, fontsize=14)
        if i % ncol == 0:
            plt.ylabel(ysign, fontsize=14)
```

## Реализация градиентного спуска

#### 5. [4 балла] Реализуйте модель линейной регрессии для функции потерь Huber loss, обучаемую градиентным спуском:

$$
L_\delta(a,y)=
\begin{cases}
 \frac{1}{2}(y - a)^2,                   & |y - a| \le \delta, \\
 \delta\, |y - a| - \frac{1}{2}\delta^2 & \textrm{иначе.}
\end{cases}
$$

Все вычисления должны быть векторизованы, циклы средствами python допускается использовать только для итераций градиентного спуска. В качестве критерия останова необходимо использовать (одновременно):

* проверку на евклидовую норму разности весов на двух соседних итерациях (например, меньше некоторого малого числа порядка $10^{-6}$, задаваемого параметром `tolerance`);
* достижение максимального числа итераций (например, 10000, задаваемого параметром `max_iter`).

Необходимо реализовать метод полного и стохастического градиентных спусков, а также поддержать метод momentum при помощи параметра `alpha` (способ оценивания градиента должен задаваться при помощи параметра `gd_type`).

Чтобы проследить, что оптимизационный процесс действительно сходится, будем использовать атрибут класса `loss_history` — в нём после вызова метода `fit` должны содержаться значения функции потерь для всех итераций, начиная с первой (до совершения первого шага по антиградиенту).

Инициализировать веса можно случайным образом или нулевым вектором. Ниже приведён шаблон класса, который должен содержать код реализации модели.

```python
from sklearn.base import BaseEstimator
import numpy as np


class HuberReg(BaseEstimator):
    def __init__(self, delta=1.0, gd_type='stochastic',
                 tolerance=1e-4, max_iter=1000, w0=None, alpha=1e-3, eta=1e-2):
        """
        gd_type: 'full' or 'stochastic'
        tolerance: for stopping gradient descent
        max_iter: maximum number of steps in gradient descent
        w0: np.array of shape (d) - init weights
        eta: learning rate
        alpha: momentum coefficient
        """
        self.delta = delta
        self.gd_type = gd_type
        self.tolerance = tolerance
        self.max_iter = max_iter
        self.w0 = w0
        self.alpha = alpha
        self.w = None
        self.eta = eta
        # list of loss function values at each training iteration
        self.loss_history = None

    def fit(self, X, y):
        """
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: self
        """
        self.loss_history = []
        self.w0 = np.zeros(X.shape[1])
        cur_w = self.w0
        previous_step_size = 1
        iter_counter = 0
        w_change = self.w0
        while previous_step_size > self.tolerance and iter_counter < self.max_iter:
            self.w = np.copy(cur_w)
            cur_w -= self.eta * self.calc_gradient(X, y) - self.alpha * w_change
            w_change = cur_w - self.w
            previous_step_size = np.linalg.norm(w_change)
            self.loss_history = np.append(self.loss_history, self.calc_loss(X, y))
            iter_counter += 1
        self.w = np.copy(cur_w)
        return self

    def predict(self, X):
        if self.w is None:
            raise Exception('Not trained yet')
        return np.matmul(X, self.w)

    def score(self, X, y):
        return 1 - ((((y - self.predict(X)) ** 2).sum()) / (((y - self.predict(X).mean()) ** 2).sum()))
    
    def huber_derivate(self, z):
        return np.where(np.abs(z) <= self.delta, z, self.delta * np.sign(z))

    def calc_gradient(self, X, y):
        """
        X: np.array of shape (l, d) (l can be equal to 1 if stochastic)
        y: np.array of shape (l)
        ---
        output: np.array of shape (d)
        """
        if self.w is None:
            raise Exception('Not trained yet')
        if self.gd_type == "full":
            z = y - np.matmul(X, self.w.T)
            N = z.shape[0]
            return -np.matmul(self.huber_derivate(z), X) / N
        else:
            i_k = np.random.randint(X.shape[0])
            return (-X[i_k] *
                    self.huber_derivate(y[i_k] -
                                        np.dot(X[i_k], self.w)))
        
    def huber_loss(self, z):
        N = z.shape[0]
        loss = np.where(np.abs(z) <= self.delta, 0.5 * ((z) ** 2),
                        self.delta * np.abs(z) - 0.5 * (self.delta ** 2))
        return np.sum(loss) / N

    def calc_loss(self, X, y):
        """
        X: np.array of shape (l, d)
        y: np.array of shape (l)
        ---
        output: float
        """
        if self.w is None:
            raise Exception('Not trained yet')
        return float(self.huber_loss(y - np.matmul(X, self.w.T)))
```

```python
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(x_train, y_train)
mm_scaled_x_train = minmax_scaler.transform(x_train)
mm_scaled_x_test = minmax_scaler.transform(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

scores_s = []
scores_f = []
Cs = np.linspace(0, 1.4)
for i in Cs:
    m_huber_i = HuberReg(alpha=i)
    m_huber_i.fit(mm_scaled_x_train, y_train)
    scores_s.append(m_huber_i.score(mm_scaled_x_test, y_test))
    m_huber_if = HuberReg(alpha=i)
    m_huber_if.fit(mm_scaled_x_train, y_train)
    scores_f.append(m_huber_if.score(mm_scaled_x_test, y_test))    

fig, ax = plt.subplots()
plt.plot(Cs, scores_s, color="C0", label="stohastic")
plt.plot(Cs, scores_f, color="C1", label="full")
plt.grid(linewidth=1)
plt.subplots_adjust(right=2, bottom=0.2, top=1.5, hspace=0.6)
plt.title("Зависимость метрики R2 от значений alpha для minmax-scaled stohastic Hubert регрессии",
          fontsize=14, weight='bold')
plt.xlabel('Значения alpha', fontsize=14)
plt.ylabel('Значение R2', fontsize=14)
plt.xticks(Cs, fontsize=12, rotation=90)
plt.legend()
plt.show()
```

### 2. Исследуем зависимость R2 от значения max_iter:

```python
standart_scaler = StandardScaler()
standart_scaler.fit(x_train, y_train)
scaled_x_train = standart_scaler.transform(x_train)
scaled_x_test = standart_scaler.transform(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

scores_s = []
scores_f = []
Cs = range(0, 10000, 250)
for i in Cs:
    s_huber_i = HuberReg(max_iter=i, delta=3, eta=1e-1)
    s_huber_i.fit(scaled_x_train, y_train)
    scores_s.append(s_huber_i.score(scaled_x_test, y_test))
    s_huber_if = HuberReg(gd_type="full", max_iter=i, delta=3, eta=1e-1)
    s_huber_if.fit(scaled_x_train, y_train)
    scores_f.append(s_huber_if.score(scaled_x_test, y_test))

fig, ax = plt.subplots()
plt.plot(Cs, scores_s, color="C0", label="stohastic")
plt.plot(Cs, scores_f, color="C1", label="full")
plt.grid(linewidth=1)
plt.subplots_adjust(right=2, bottom=0.2, top=1.5, hspace=0.6)
plt.title("Зависимость метрики R2 от значений max_iter для standard-scaled Hubert регрессии",
          fontsize=14, weight='bold')
plt.xlabel('Значения max_iter', fontsize=14)
plt.ylabel('Значение R2', fontsize=14)
plt.xticks(Cs, fontsize=12, rotation=90)
plt.yticks(range(-8,2,1))
plt.legend()
plt.show()
```

Видно, что stohastic standart-scaled также даёт неустойчивый метод. А full не изменятся при изменении только max_iter (причём, если не ставить параметры для eta и delta, то значение R2 остаётся 0). При таких же значениях оно равно 0.113...

```python
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(x_train, y_train)
mm_scaled_x_train = minmax_scaler.transform(x_train)
mm_scaled_x_test = minmax_scaler.transform(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

scores_s = []
scores_f = []
Cs = range(0, 10000, 250)
for i in Cs:
    m_huber_i = HuberReg(max_iter=i)
    m_huber_i.fit(mm_scaled_x_train, y_train)
    scores_s.append(m_huber_i.score(mm_scaled_x_test, y_test))
    m_huber_if = HuberReg(max_iter=i)
    m_huber_if.fit(mm_scaled_x_train, y_train)
    scores_f.append(m_huber_if.score(mm_scaled_x_test, y_test))
    
fig, ax = plt.subplots()
plt.plot(Cs, scores_s, color="C0", label="stohastic")
plt.plot(Cs, scores_f, color="C1", label="full")
plt.grid(linewidth=1)
plt.subplots_adjust(right=2, bottom=0.2, top=1.5, hspace=0.6)
plt.title("Зависимость метрики R2 от значений max_iter для minmax-scaled Hubert регрессии",
          fontsize=14, weight='bold')
plt.xlabel('Значения max_iter', fontsize=14)
plt.ylabel('Значение R2', fontsize=14)
plt.xticks(Cs, fontsize=12, rotation=90)
plt.legend()
plt.show()
```

Изменения max_iter влияют неустойчиво на алгоритм, как в случае использования full, так и stohastic. И это было вполне ожидаемо

Сделайте выводы о скорости сходимости различных модификаций градиентного спуска.

#### Построю графики зависимости для метода итераций при фиксированных других параметрах, которые были найдены из предыдущего задания (alpha=0.985, max_iter=1000) и буду масштабировать MinMaxScaler. Для остальных алгоритмов также возьму max_iter=1000 (соответственно, alpha=0)

```python
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(x_train, y_train)
mm_scaled_x_train = minmax_scaler.transform(x_train)
mm_scaled_x_test = minmax_scaler.transform(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

m_huber_stoh = HuberReg(alpha=0)
m_huber_stoh.fit(mm_scaled_x_train, y_train)

m_huber_full = HuberReg(alpha=0)
m_huber_full.fit(mm_scaled_x_train, y_train)

m_huber_full_momentum = HuberReg(alpha=0.985)
m_huber_full_momentum.fit(mm_scaled_x_train, y_train)

fig, ax = plt.subplots()
plt.plot(range(1, len(m_huber_stoh.loss_history) + 1),
         m_huber_stoh.loss_history,
         color="C0", label="stohastic")
plt.plot(range(1, len(m_huber_full.loss_history) + 1),
         m_huber_full.loss_history,
         color="C1", label="full")
plt.plot(range(1, len(m_huber_full_momentum.loss_history) + 1),
         m_huber_full_momentum.loss_history, color="C2", label="full with momentum")
plt.grid(linewidth=1)
plt.subplots_adjust(right=2, bottom=0.2, top=1.5, hspace=0.6)
plt.title("Зависимость функции потерь от итерации для различных видов Hubert регрессии",
          fontsize=14, weight='bold')
plt.xlabel('Номер итерации', fontsize=14)
plt.ylabel('Значение функции потерь', fontsize=14)
scale = max(len(m_huber_full.loss_history) + 1,
            len(m_huber_stoh.loss_history) + 1,
            len(m_huber_full_momentum.loss_history) + 1)
plt.xticks(range(1, scale + 1, 100), fontsize=12)
plt.legend()
plt.show()
```

#### Соответственно, если варьировать eta, то можно получить необходимую точность и для алгоритмов с alpha=0:

```python
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(x_train, y_train)
mm_scaled_x_train = minmax_scaler.transform(x_train)
mm_scaled_x_test = minmax_scaler.transform(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

m_huber_stoh = HuberReg(alpha=0, eta=0.2)
m_huber_stoh.fit(mm_scaled_x_train, y_train)

m_huber_full = HuberReg(alpha=0, eta=0.2)
m_huber_full.fit(mm_scaled_x_train, y_train)

m_huber_full_momentum = HuberReg(alpha=0.985)
m_huber_full_momentum.fit(mm_scaled_x_train, y_train)

fig, ax = plt.subplots()
plt.plot(range(1, len(m_huber_stoh.loss_history) + 1),
         m_huber_stoh.loss_history,
         color="C0", label="stohastic")
plt.plot(range(1, len(m_huber_full.loss_history) + 1),
         m_huber_full.loss_history,
         color="C1", label="full")
plt.plot(range(1, len(m_huber_full_momentum.loss_history) + 1),
         m_huber_full_momentum.loss_history, color="C2", label="full with momentum")
plt.grid(linewidth=1)
plt.subplots_adjust(right=2, bottom=0.2, top=1.5, hspace=0.6)
plt.title("Зависимость функции потерь от итерации для различных видов Hubert регрессии",
          fontsize=14, weight='bold')
plt.xlabel('Номер итерации', fontsize=14)
plt.ylabel('Значение функции потерь', fontsize=14)
scale = max(len(m_huber_full.loss_history) + 1,
            len(m_huber_stoh.loss_history) + 1,
            len(m_huber_full_momentum.loss_history) + 1)
plt.xticks(range(1, scale + 1, 100), fontsize=12)
plt.legend()
plt.show()
```

Логичным выводом из графика будет следующий: Градиентные алгоритмы без метода momentum сходятся за меньшее число итераций (останавливаясь по критерию сравнения нормы с порогом), но дают более плохой (в плане метрики R2) результат. В то время как полный градиентный спуск сходится долще, но даёт значительно лучший результат по метрике R2.

Хотелось бы добавить, что на точность (по R2) алгоритмов с alpha=0, порой, сильно влияло разбиение выборки.
