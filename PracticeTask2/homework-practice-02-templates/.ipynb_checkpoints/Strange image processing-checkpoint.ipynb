```python
import numpy as np
import skimage.io as io
import skimage.color as color
import scipy.misc as misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = misc.imread('msu6.jpg')
```

```python
plt.imshow(img)
```

```python
g_i = np.sum(img * np.array([0.299, 0.587, 0.114]), axis=2)
```

```python
g_i
```

```python
gr_i = color.rgb2grey(img)
```

```python
gr_i
```

```python
plt.imshow(gr_i, cmap=plt.cm.Greys)
```
