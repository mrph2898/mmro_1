## 2. Сравнение pairwise_distance с scipy.spatial.distance.cdist:

```python
import functions as f0
import functions_vectorized as f1
import time
import scipy.spatial.distance as dist
N = 1000
```

```python
d = 50 ##Почему 50 - размер признакового пространства? - см. в сравнении двух реализаций
```

```python
time_values0 = np.empty(N)
time_values1 = np.empty(N)
time_values2 = np.empty(N)
for i in range(0, N): 
    x = np.random.randn(i, d)
    y = np.random.randn(i, d)
    start = time.time()
    f0.pairwise_distance(x, y)
    time_values0[i] = time.time() - start
    
    start = time.time()
    f1.pairwise_distance(x, y)
    time_values1[i] = time.time() - start
    
    start = time.time()
    dist.cdist(x, y)
    time_values2[i] = time.time() - start
```

```python
fig, ax = plt.subplots()

ax.plot(range(0, N), time_values0, color='C0', label='Невекторизованный')
ax.plot(range(0, N), time_values1, color='C1', label='Векторизованный')
ax.plot(range(0, N), time_values2, color='C2', label='scipy.spatial.distance.cdist')
plt.grid(linewidth=1)
plt.subplots_adjust(right=2, bottom=0.2, top=1.5, hspace=0.6)
plt.title("Сравнение векторизованого и невекторизованного pairwise_distance c scipy.spatial.distance.cdist по размеру axis=0(объёму выборки)", fontsize=14, weight='bold')
plt.xlabel('Размер axis=0(объём выборки)', fontsize=14)
plt.ylabel('Время[сек]', fontsize=14)
plt.xticks(range(0, N + 50, 50))
plt.yticks(fontsize=12)
plt.yscale('log') 
plt.legend()
plt.show()
```

### Однозначно, алгоритм scipy.spatial.distance.cdist работает быстрее моих реализаций
